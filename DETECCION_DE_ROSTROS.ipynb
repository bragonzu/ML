```python
import cv2 #Importar libreria opencv
import numpy as np
import imutils
import os
from google.colab.patches import cv2_imshow
```

Cargamos la imagen

```python
img=cv2.imread("/content/drive/MyDrive/people.jpg")
#Examples
#img=cv2.imread("/content/drive/MyDrive/people2.png")
#img=cv2.imread("/content/drive/MyDrive/people3.jpg")
#img=cv2.imread("/content/drive/MyDrive/people4.jpg")
#Escala de grises gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

Importamos XML 

```python
cascada_faces=cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
```

Coordenamos los rostros 


```python
coord_faces=cascada_faces.detectMultiScale(img,
                                           scaleFactor= 1.08,
                                           minNeighbors=5,
                                           minSize=(30,30),
                                           maxSize=(200,200))
```

Recorrido for para coleccionar los rostros


```python
for (x,y,w,h) in coord_faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

Muestra de resultado


```python
cv2_imshow(img)
```
